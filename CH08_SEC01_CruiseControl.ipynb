```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams.update({'font.size': 15})
plt.rcParams['figure.figsize'] = [12, 12]
```

```python
t = np.arange(0,10,0.01)   # time

idealSpeed = 60 * np.ones_like(t)  # reference speed
disturbance = 10*np.sin(np.pi*t)     

wishfulConstant = 1                
actualConstant = 0.5                         

#open loop control law: u = y
gasFedIntoOpenLoop = idealSpeed*wishfulConstant      # I want my gas input to equal speed       

responseOfOpenLoop = actualConstant * gasFedIntoOpenLoop + disturbance #(+/- 10)
print("The speed in a open loop system is in the range of 20 -40 mph")


#               closed loop control law: u = k(w - y)
#               actualConstant =  k(idealSpeed - responseOfCloseLoop)         equation(8.6 )        
# control gain, actualConstant = 2K(idealspeed - responseOfCloseLoop)
#and on isolating y, you get y = 2K / (2K + 1)*idealSpeed                      plus add some disturbance (feedback) 

K= 50
responseOfCloseLoop = (actualConstant*K/(1+actualConstant*K))*idealSpeed + disturbance/(1+actualConstant*K) #(+/-10 / 26) < 1
print("The speed in a closed loop system is in the range of 58 - 59 mph")
```

```python
plt.plot(t,idealSpeed,'k',linewidth=2,label='Ideal Speed')
plt.plot(t,disturbance,'k--',linewidth=1.5,label='Disturbance')
plt.plot(t,responseOfOpenLoop,'r',linewidth=1.5,label='Open Loop')
plt.plot(t,responseOfCloseLoop,'b',linewidth=1.5,label='Closed Loop')

plt.xlabel('Time')
plt.ylabel('Speed')

plt.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```
